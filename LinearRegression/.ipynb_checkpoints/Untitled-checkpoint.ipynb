```python
data = pd.read_csv('./Advertising.csv',index_col=0) # csv文件第一列作为dataframe的index
data.head()
```

## 要线性回归的对象

```python
sns.scatterplot(data.TV,data.sales)
```

```python
X_train = data.TV.values
y_train = data.sales.values
X_train.shape
y_train.shape
```

## 构造模型

```python
class LinearRegressionModel(tf.keras.Model):
    def __init__(self,):
        super().__init__()
        self.w = tf.Variable(tf.random.normal((1,1)))
        self.b = tf.Variable(tf.random.normal((1,1)))
        
        
    def call(self,input):
        output = self.w*input + self.b
        return output
```

```python
# 以下代码结构与前节类似
model = LinearRegressionModel()
```

```python
optimizer = tf.keras.optimizers.Adam(learning_rate=0.01)
for i in range(1000):
    with tf.GradientTape() as tape:
        y_pred = model(X_train)      # 调用模型 y_pred = model(X) 而不是显式写出 y_pred = a * X + b
        loss = tf.reduce_mean(tf.square(y_pred - y_train))
    grads = tape.gradient(loss, model.variables)    # 使用 model.variables 这一属性直接获得模型中的所有变量
    optimizer.apply_gradients(grads_and_vars=zip(grads, model.variables))
print(model.variables[0].numpy())
```

```python
sns.scatterplot(data.TV,data.sales)
sns.lineplot(X_train,model.variables[0].numpy().flatten()*X_train+model.variables[1].numpy().flatten(),color='red')
```

```python
from mpl_toolkits import mplot3d

def f(w, b):
    out = []
    out_array = np.zeros((len(w),len(b)))
    for x_index,x in enumerate(w):
        for y_index,y in enumerate(b):
            out_array[x_index][y_index] = tf.reduce_mean(tf.square(x*X_train+y - y_train)).numpy()
    
    return out_array
w = np.linspace(-0.5, 0.5, 300)
b = np.linspace(-10, 10, 300)
W, B = np.meshgrid(w, b)
Z = f(w, b)
Z
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_wireframe(W, B, Z, cmap='viridis')
ax.scatter(1,2,)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```
